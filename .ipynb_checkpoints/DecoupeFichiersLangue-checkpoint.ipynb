```python
import codecs
```

```python
def getFichierTag(numLangue):
    nFichierTag="/Users/gilles/ownCloud/Cours/Bordeaux/L3-Morphologie3/Lot1/Groupes/Extraits-Lot1-%s-TG.text"%numLangue
    with codecs.open(nFichierTag,"r",encoding="utf8") as inFile:
        lignes=[l.strip() for l in inFile.readlines()]

    freqMots={}
    for l in lignes:
        mot,cat,lemme=l.split("\t")
        if cat in ["NOM","VER","ADJ","ADV"] and not lemme in ["<unknown>","ne","pas"]:
            if not lemme in freqMots:
                freqMots[lemme]=0
            freqMots[lemme]+=1
    return freqMots
```

```python
for l in ["%02d"%(n+1) for n in range(11)]:
    print "Langue",l
    print "--------------"
    freqMots=getFichierTag(l)
    for k in sorted(freqMots, key=freqMots.get,reverse=True):
        if freqMots[k]>1: 
            print k,freqMots[k]
    print
    print "================================="
    print
```
