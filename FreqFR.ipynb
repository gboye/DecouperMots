```python
nbLots=2
nRepLot="/Users/gilles/ownCloud/Cours/Bordeaux/L3-Morphologie3/Lots/"
```

```python
def getFichierTag(numLangue,numLot):
    nFichierTag=nRepLot+"Extraits-%s-Lot%d-TG.text"%(numLangue,numLot)
    with codecs.open(nFichierTag,"r",encoding="utf8") as inFile:
        lignes=[l.strip() for l in inFile.readlines()]

    freqMots={}
    for l in lignes:
        mot,cat,lemme=l.split("\t")
        if (cat in ["NOM","ADJ","ADV"] or "VER" in cat) and not lemme in ["<unknown>","ne","pas"]:
            if not lemme in freqMots:
                freqMots[lemme]=0
            freqMots[lemme]+=1
    return freqMots
```
